```python

s1 = {1, 2, 3}
s2 = {1, 2, 3, 4, 5}
if s1.issubset(s2):
    print("s1 is subset of s2")
else:
    print("s1 is not subset of s2")
```

```python

A = {2, 4, 6}
B = {1, 3, 5}
print(A.symmetric_difference(B))
```
