```python
#WAP wich will take a number, is even it will calculate even factorial and is odd it will calculate odd factorialinpur(n)

def even_fact(n):
    ans = 1
    for i in range(2, n+1, 2):   # loop through even numbers
        ans *= i
    return ans

def odd_fact(n):
    ans = 1
    for i in range(1, n+1, 2):   # loop through odd numbers
        ans *= i
    return ans

def factors(n):
    print("Factors of", n, "are:", end=" ")
    for i in range(1, n+1):
        if n % i == 0:
            print(i, end=" ")
    print()

# main program
n = int(input("Enter a number: "))

if n % 2 == 0:
    print("Even factorial of", n, "is:", even_fact(n))
else:
    print("Odd factorial of", n, "is:", odd_fact(n))

factors(n)
     
```
