```python
lst=[[3,1,1,2],[50,4,8,10],[2,3,7,8]]
ans=[]
for i in range(len(lst)):
    new=[]
    for j in lst[i]:
        if j %2 == 0:
            new.append(j-1)
        else:
             new.append(j+1)
    ans.append(new)
print(lst)
print(ans)
           
```
