```python
def show(n):
    if n==0:
        return
    print(n)
   
    show(n-1)
    print("END")

show(5)
```

```python
# fctorial 
def fact(n):
    if (n==0 or n==1):
        return 1
    
    return fact(n-1)*n
    
print(fact(7))
```

```python
n=int(input("Enter no"))
def netural(n):
    if(n==0):
        return 0
   
    return n+netural(n-1)

print(netural(n))
```
