```python
email = input("Enter email: ")

# check conditions
if "@" in email and "." in email:
    at_pos = email.index("@")
    dot_pos = email.rfind(".")   # last dot

    if at_pos < dot_pos:
        print("Valid Email")
    else:
        print("Invalid Email")
else:
    print("Invalid Email")
```
